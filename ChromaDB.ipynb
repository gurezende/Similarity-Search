```python
#Load files
df = pd.read_json("healthnews.json", lines=True)
df.sample(3)
```

```python
# First, we need to create a client to connect to the database
client = chromadb.Client()
```

ChromaDB is the easiest vector database to work with. It is Open Source and integrates well with packages like Langchain.
Let's start by creating a new collection.
A collection are something like a table in common relational databases.
In simple terms, think of a vector database as a *library*, and the *collections* as different sections of that library, like "fiction," "non-fiction," or "reference." Each section (collection) contains related items (vectors).

```python
# Creating a collection
collection = client.create_collection("my_collection")
```

Next, let's load some data from a Pandas dataframe to add to our collection

Adding data to a collection is pretty simple

```python
# Adding data to the collection
collection.add(
    documents=df["News"].tolist(), 
    metadatas=[{"Date": str(dt)} for dt in df["Date"].tolist()],
    ids=[str(i) for i in range(len(df))]
) 
```

```python
collection.count()
```

```python
collection.get()
```

```python
d= collection.get(ids=["0"], include=['embeddings', 'metadatas'])
d['embeddings'][0][:20]
```

```python
%%time
# Text for the query
query = ["Quick workout"]

# Query the collection
results = collection.query(
    query_texts=query, # Chroma will embed this for you
    n_results=3 # how many results to return
)

results
```

```python
# Update a collection
collection.upsert(
    documents=df["News"].tolist(), 
    metadatas=[{"Date": str(dt)} for dt in df["Date"].tolist()],
    ids=[str(i) for i in range(len(df))]
)
```
